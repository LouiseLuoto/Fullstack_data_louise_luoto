 f) Graph different data engineers salary. There are many approaches to graph this, try different and reason about pros and cons of each you choose.

```python
position_salaries = df.loc[df["Position"].isin(["Junior Data Engineer", 
                                                "Senior Data Engineer", 
                                                "Big Data Engineer"]),
                           ["Position", "Salary_SEK"]]
position_salaries
```

```python
import matplotlib.pyplot as plt

position_salaries = position_salaries.sort_values(by=["Position", "Salary_SEK"])

plt.figure(figsize=(10, 6))

for position in position_salaries['Position'].unique():
    subset = position_salaries[position_salaries['Position'] == position]
    plt.plot(subset['Salary_SEK'].values, marker='o', label=position)

plt.title('Salaries for Different Data Engineer Positions')
plt.ylabel('Salary in SEK')
plt.legend(title='Position')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

positions = ["Junior Data Engineer", "Senior Data Engineer", "Big Data Engineer"]
colors = ["orange", "green", "blue"]

for i, position in enumerate(positions):
    subset = position_salaries[position_salaries["Position"] == position]
    axes[i].bar(range(len(subset)), subset["Salary_SEK"], color=colors[i])
    axes[i].set_title(f"Salary for {position}")
    axes[i].set_ylabel("Salary in SEK")

plt.tight_layout()
plt.show()
```
